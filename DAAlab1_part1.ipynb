```python
import time
import matplotlib.pyplot as plt

n=[3,4,5,6,9]
timesIter=[]
timesRecur=[]

st=time.time()
sum=0
#iterative method
for k in n:
  for i in range(1,k+1):
    sum=sum+i
  et=time.time()
  elapsed=et-st
  timesIter.append(elapsed)


#recursive method
sum=0
def sumofn(n):
  if(n <= 1):
    return n
  return n+sumofn(n-1)

for k in n:
  st=time.time()
  sumofn(k)
  et=time.time()
  elapsed=et-st
  timesRecur.append(elapsed)

plt.bar(n,timesIter)
plt.show()
plt.bar(n,timesRecur)
plt.show()
```

```python
import random
import numpy as np
import time
import matplotlib.pyplot as plt


A=np.random.randint(1000,size=(10000))
key=int(input("enter the number you want to search"))
times=[]
algo=['linear','binary']

st=time.time()
#linear search
for i in range(0,len(A)):
  if(i == key):
    print("element found")
    break
et=time.time()
elapsed=et-st
times.append(elapsed)

#binary
n=len(A)
l=A[0]
r=A[n-1]
mid=0
while(l <= r):
  mid= l + (r-1) // 2
  if (A[mid]== key):
    print("element found")
  elif mid < key:
    l = mid + 1
  else:
    r = mid - 1
```
